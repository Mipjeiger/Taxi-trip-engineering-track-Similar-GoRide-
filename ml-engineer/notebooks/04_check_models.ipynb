```python
from pathlib import Path
import pickle
import joblib
import numpy as np
import tensorflow as tf

MODEL_DIR = Path.cwd().parent / "models"

for model_path in sorted(MODEL_DIR.iterdir()):

    # ── .pkl files ───────────────────────────────────────────────────────────
    if model_path.suffix == ".pkl":
        print(f"\n{'='*50}")
        print(f"📦 {model_path.name}")

        # Try pickle first, fallback to joblib
        try:
            with open(model_path, "rb") as f:
                obj = pickle.load(f)
            print(f"  Loader     : pickle")
        except Exception:
            try:
                obj = joblib.load(model_path)
                print(f"  Loader     : joblib")
            except Exception as e:
                print(f"  ❌ Failed to load: {e}")
                continue

        print(f"  Type       : {type(obj)}")

        if isinstance(obj, (list, np.ndarray)):
            print(f"  Length     : {len(obj)}")
            print(f"  Contents   : {obj}")

        elif isinstance(obj, dict):
            print(f"  Keys       : {list(obj.keys())}")
            for k, v in obj.items():
                print(f"    [{k}] → {type(v).__name__}")
                if hasattr(v, "get_params"):
                    print(f"           params: {v.get_params()}")

        elif hasattr(obj, "get_params"):
            print(f"  Params     : {obj.get_params()}")

        if hasattr(obj, "classes_"):
            print(f"  Classes    : {obj.classes_}")

        if hasattr(obj, "mean_"):
            print(f"  Mean       : {obj.mean_}")
            print(f"  Scale      : {obj.scale_}")
            print(f"  N features : {obj.n_features_in_}")

    # ── .h5 files ────────────────────────────────────────────────────────────
    elif model_path.suffix == ".h5":
        print(f"\n{'='*50}")
        print(f"🧠 {model_path.name}")
        try:
            model = tf.keras.models.load_model(model_path)
            model.summary()
            print(f"  Input  shape : {model.input_shape}")
            print(f"  Output shape : {model.output_shape}")
            print(f"  Total params : {model.count_params():,}")
        except Exception as e:
            print(f"  ❌ Failed to load: {e}")
```

```python
# List features
for model_path in sorted(MODEL_DIR.iterdir()):
    if model_path.suffix == ".pkl":

        try:
            with open(model_path, "rb") as f:
                obj = pickle.load(f)
        except Exception:
            try:
                obj = joblib.load(model_path)
            except Exception:
                continue

        if hasattr(obj, "feature_names_in_"):
            print(f" {model_path.name} features: {obj.feature_names_in_}")
```
