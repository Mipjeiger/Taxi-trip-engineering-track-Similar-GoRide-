## Google maps

```python
from IPython.display import IFrame

def show_route(df, idx=0):
    pickup = df.loc[idx, "Pickup Location"]
    drop = df.loc[idx, "Drop Location"]

    url = f"https://www.google.com/maps?q={pickup.replace(' ', '+')}+to+{drop.replace(' ', '+')}&output=embed"

    return IFrame(url, width=1100, height=600)

show_route(df, idx=0)
```

## Multiple ride maps

```python
from geopy.geocoders import Nominatim
import time

geolocator = Nominatim(user_agent="ride_app")

def get_lat_lon(location):
    try:
        loc = geolocator.geocode(location)
        time.sleep(1)  # To respect the usage policy of Nominatim
        return loc.latitude, loc.longitude
    except:
        return None, None
```

## Function for distance pickup location - drop location estimation on google maps

```python
df_sample = df.head(20).copy()
df_sample["pickup_lat"], df_sample["pickup_lon"] = zip(*df_sample["Pickup Location"].apply(get_lat_lon))
df_sample["drop_lat"], df_sample["drop_lon"] = zip(*df_sample["Drop Location"].apply(get_lat_lon))
df_sample.head()
```

```python
# Interactive Map with Folium (Direct Display)
import folium
from folium import plugins
import pandas as pd

# Calculate center coordinates
center_lat = df_sample["pickup_lat"].mean()
center_lon = df_sample["pickup_lon"].mean()

# Create folium map
ride_map = folium.Map(
    location=[center_lat, center_lon],
    zoom_start=12,
    tiles='OpenStreetMap'
)

# Plot all routes
for idx, (_, row) in enumerate(df_sample.head(20).iterrows()):
    if all([
        row["pickup_lat"] is not None,
        row["pickup_lon"] is not None,
        row["drop_lat"] is not None,
        row["drop_lon"] is not None,
        pd.notna(row["pickup_lat"]),
        pd.notna(row["drop_lat"])
    ]):
        try:
            # Pickup marker (green)
            folium.Marker(
                location=[row["pickup_lat"], row["pickup_lon"]],
                popup=f"<b>Pickup {idx+1}</b><br>Lat: {row['pickup_lat']:.4f}<br>Lon: {row['pickup_lon']:.4f}",
                tooltip=f"Pickup {idx+1}",
                icon=folium.Icon(color='green', icon='play', icon_color='white')
            ).add_to(ride_map)
            
            # Drop marker (red)
            folium.Marker(
                location=[row["drop_lat"], row["drop_lon"]],
                popup=f"<b>Drop {idx+1}</b><br>Lat: {row['drop_lat']:.4f}<br>Lon: {row['drop_lon']:.4f}",
                tooltip=f"Drop {idx+1}",
                icon=folium.Icon(color='red', icon='stop', icon_color='white')
            ).add_to(ride_map)
            
            # Draw route line
            folium.PolyLine(
                locations=[
                    [row["pickup_lat"], row["pickup_lon"]],
                    [row["drop_lat"], row["drop_lon"]]
                ],
                color='blue',
                weight=2,
                opacity=0.7,
                popup=f"Route {idx+1}"
            ).add_to(ride_map)
            
        except Exception as e:
            print(f"Error plotting route {idx+1}: {e}")

# Add title/legend
title_html = '''
             <div style="position: fixed; 
                     top: 10px; left: 35px; width: 180px; height: 120px; 
                     background-color: white; border:2px solid grey; z-index:9999; 
                     font-size:14px; padding: 10px; border-radius: 5px;">
             <b style="font-size: 16px;">Routes Overview</b><br>
             🟢 Green: Pickup Location<br>
             🔴 Red: Drop Location<br>
             🔵 Blue: Ride Route<br>
             <b>Total Routes:</b> {}</div>
             '''.format(len(df_sample))

ride_map.get_root().html.add_child(folium.Element(title_html))

print(f"✓ Interactive map generated with {len(df_sample)} routes")
print(f"✓ Map center: ({center_lat:.4f}, {center_lon:.4f})")

# Display directly
ride_map
```

## Create route effience

```python
df["Efficiency_routes"] = df["Ride Distance"] / df["Avg CTAT"] # Efficiency = Distance / Time average speed (km/h)

top_routes = (
    df.groupby(["Pickup Location", "Drop Location"])
    .agg({
        "Efficiency_routes": "mean",
        "Avg CTAT": "mean",
        "Ride Distance": "mean",
        "route_count": "mean"
    })
    .sort_values(by="Efficiency_routes", ascending=False).head(20)
)
top_routes
```

```python
# Create plot for efficiency routes (simple approach)
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
route_names = [f"{pickup} → {drop}" for pickup, drop in top_routes.index]
plt.barh(route_names, top_routes["Efficiency_routes"], color='skyblue')
plt.title("Top 20 Most Efficient Routes (Distance/Time)", fontsize=16)
plt.xlabel("Efficiency (km/h)", fontsize=14)
plt.ylabel("Route (Pickup → Drop)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Create traffic heatmap
from folium.plugins import HeatMap

m = folium.Map(location=[center_lat, center_lon], zoom_start=12)
heat_data = df_sample[["pickup_lat", "pickup_lon"]].dropna().values.tolist()
HeatMap(heat_data).add_to(m)
m
```
